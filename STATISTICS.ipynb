```python
import numpy as np
##sales data
lokesh=np.array([50,54,48,52,45])
mohan=np.array([60,70,90,40,65])
#mean
print("mean value is:",np.mean(lokesh))
print("mean value is;",np.mean(mohan))
##variance
print("variance is:",np.var(lokesh))
print("variance is:",np.var(mohan))
##standard deviation
print("standard deviation is:",np.std(lokesh))
print("standard deviation is:",np.std(mohan))
##IQR
lokesh_iqr=np.percentile(lokesh,75)-np.percentile(lokesh,25)
mohan_iqr=np.percentile(mohan,75)-np.percentile(mohan,25)
print("lokesh IQR IS:",lokesh_iqr)
print("mohan_IQR IS:",mohan_iqr)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew,kurtosis
#Employee monthly sales (in thousands)
sales=np.array([40,42,45,50,52,55,58,60,65,70,150,200])
#Skewness & Kurtosis
sales_skew=skew(sales)
sales_kurt=kurtosis(sales) #by default Fisher's definition (Normal = 0)
                           # Add +3 if you want pearson defintion
print(f"skewness: {sales_skew:.2f}")
print(f"kurtosis (Fisher): {sales_kurt:.2f}")
print(f"kurtosis (Pearson):{sales_kurt+3:.2f}")
#plot Histogram
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(sales,bins=8,color="skyblue",edgecolor="black")
plt.title("Employee sales Distribution")
plt.xlabel("sales(in 000s)")
plt.ylabel("Frequency")
#Boxplot
plt.subplot(1,2,2)
plt.boxplot(sales, vert=False)
plt.title("Boxplot of sales")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew,kurtosis

#Cricket runs scored in different matches
lokesh_runs=np.array([25,30,35,40,45,50,55,100,120,150])
mohan_runs=np.array([40,42,45,47,50,52,53,55,56,58])

# Skewness & Kurtosis
print("lokesh:")
print(f"skewness: {skew(lokesh_runs):.2f}")
print(f"kurtosis(Fisher): {kurtosis(lokesh_runs):.2f}")
print(f"kurtosis(Pearson): {kurtosis(lokesh_runs)+3:.2f}\n")

print("mohan:")
print(f"skewness: {skew(mohan_runs):.2f}")
print(f"kurtosis(Fisher): {kurtosis(mohan_runs):.2f}")
print(f"kurtosis(Pearson): {kurtosis(mohan_runs)+3:.2f}")

#Plotting
plt.figure(figsize=(12,6))

#Histogram comparsion
plt.subplot(2,2,1)
plt.hist(lokesh_runs,bins=6,color="Orange",edgecolor="black")
plt.title("Lokesh_Runs Histogram")
plt.xlabel("Runs")
plt.ylabel("Frequency")

plt.subplot(2,2,2)
plt.hist(mohan_runs,bins=6,color="skyblue",edgecolor="black")
plt.title("Mohan_Runs Histogram")
plt.xlabel("Runs")
plt.ylabel("Frequency")

#Boxplot comparsion
plt.subplot(2,1,2)
plt.boxplot([lokesh_runs,mohan_runs],labels=["lokesh","mohan"])
plt.title("Boxplot of Runs")
plt.ylabel("Runs")
plt.tight_layout()
plt.show()


```
